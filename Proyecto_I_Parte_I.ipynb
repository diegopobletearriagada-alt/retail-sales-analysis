2.Calcula el promedio de ventas diarias por categoría de producto.

```python
def calcular_promedio_diario_por_categoria(datos):
    #  Sacar categorías únicas
    categorias = np.unique(datos[:, 5])

    #Diccionario DONDE SE guardara el promedio por categoría:
    promedio_diario_por_categoria = {}

    #  Recorrer cada categoría
    for categoria in categorias:
        filas_categoria = datos[datos[:, 5] == categoria]      # VAR LOCAL A:filas de esa categoría

        total_categoria = filas_categoria[:, 8].astype(float).sum()   # VAR LOCALB: total ventas de esa categoría.
        cantidad_dias = len(filas_categoria) # VAR LOCALC:cuántos registros (días) hay

        promedio = total_categoria / cantidad_dias    # VAR LOCAL (D): total/cantidad de días

        promedio_diario_por_categoria[categoria] = promedio            # guardar en el diccionario

    return promedio_diario_por_categoria
```

```python
promedios = calcular_promedio_diario_por_categoria(datos)

print("Promedio de ventas diarias por categoría:")
for categoria, promedio in promedios.items():
    print(f"Categoría: {categoria} --- Promedio diario: ${promedio:,.2f}")
```

3.Identifica las categorías de productos con mayores y menores ventas.

```python
def encontrar_categoria_mayor_menor_ventas(totales_por_categoria): ## Esta función recibirá un ARRAY con los totales de ventas por categoría
    # Encuentra la categoría con mayor y menor ventas totales
    categoria_mayor_ventas = np.argmax(totales_por_categoria) ## me devuelve evuelve la POSICIÓN donde el arrayy tiene el valor más grande.
    categoria_menor_ventas = np.argmin(totales_por_categoria)
    return categoria_mayor_ventas, categoria_menor_ventas
```
